# Adding one additional attribute using Fetaure Engineering i.e. Depreciation Rate

```python
len(data['new-price'].unique())

price_range=[]
for i in data['new-price']:
    price_range.append(i.split(" ")[5])

price_range_numbers=[]
for i in price_range:
    price_range_numbers.append(i.split("Rs")[1][1:])

initial_price=[]
final_price=[]
for i in price_range_numbers:
    l=(i.split('-'))
    if(len(l)>1):
        initial_price.append(l[0])
        final_price.append(l[1])
    else:
        initial_price.append(l[0])
        final_price.append(0)

len(initial_price)
len(final_price)

initial_price_number=[]

    
import re
new_l=[]
for i in initial_price:
    new_l.append(i.replace("Cr"," "))
    
only_numbers_starting_range=[]
for i in new_l:
    only_numbers_starting_range.append(float(i))
    
only_nubers_ending_range=[]
for i in final_price:
    only_nubers_ending_range.append(float(i))
    
#Lets calculate average price and substract it from the current price and mark it as depreciation price

average_price=[]
total_records=len(only_nubers_ending_range)

for i in range(total_records):
    average_price.append(abs(only_nubers_ending_range[i]-only_numbers_starting_range[i]))
    
data['deprication_rate']=average_price

data.info()


#Lets go ahead with the ensemble techniques:
#RAndom Forest 


    
```
